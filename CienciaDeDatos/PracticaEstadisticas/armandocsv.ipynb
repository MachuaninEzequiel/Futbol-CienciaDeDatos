```python
partido_posiciones = partido[['player', 'location', 'carry_end_location',  'possession', 'timestamp', 'type', 'second', 'play_pattern', 'period', 'minute', 'team']]
partido_posiciones = partido_posiciones.dropna()
partido_posiciones
```

```python
partido_posiciones = partido_posiciones.reset_index(drop=True)
posicionesArg = partido_posiciones[partido_posiciones['team'] == 'Argentina']
posicionesCol = partido_posiciones[partido_posiciones['team'] == 'Colombia']

```

```python
df_argentina = posicionesArg[['player', 'location', 'carry_end_location', 'timestamp', 'type', 'second', 'minute']]
df_argentina_grouped = df_argentina.groupby('player').agg(list).reset_index()
df_argentina_grouped
```

```python
# Pivotar el dataframe de Argentina
pivot_argentina = df_argentina.pivot_table(index=['timestamp', 'type', 'minute', 'second'], columns='player', values=['location'], aggfunc=lambda x: ' '.join(map(str, x)))
pivot_argentina = pivot_argentina.reset_index()
pivot_argentina_sorted = pivot_argentina.sort_values(by=['minute', 'second'])
pivot_argentina_sorted
```
