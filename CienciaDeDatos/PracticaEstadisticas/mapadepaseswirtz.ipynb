```python
from statsbombpy import sb
import pandas as pd 
import matplotlib.pyplot as plt
from mplsoccer import  VerticalPitch, Pitch
import numpy as np
```

```python
sb.competitions()
```

```python
bundesliga = sb.matches(competition_id= 9, season_id=281).sort_values(by='match_date')
bundesliga
```

```python
partido = sb.events(match_id=3895302)
partido.player.value_counts()
```

```python
wirtz = partido[partido['player'] == 'Florian Wirtz' ]
```

```python
fig, ax = plt.subplots(figsize=(16,9))
pitch = Pitch(
    pitch_type='statsbomb',
    axis=True,
    label=True,
    tick=True,
    goal_type='box'
)

pitch.draw(ax=ax)
```

```python
paseswirtz = wirtz[wirtz['type'] == 'Pass']
paseswirtz[['x', 'y']] = paseswirtz.location.apply(pd.Series)
paseswirtz[['endx', 'endy']] = paseswirtz.pass_end_location.apply(pd.Series)
paseswirtz[['x','y','endx','endy']]
```

```python
fig, ax = plt.subplots(figsize=(16,9))
pitch = Pitch(
    pitch_type='statsbomb',
    axis=True,
    label=True,
    tick=True,
    goal_type='box'
)

pitch.draw(ax=ax)

pases_completos = paseswirtz[~paseswirtz['pass_outcome'].isin(['Incomplete', 'Pass Offside'])]
pases_incompletos = paseswirtz[paseswirtz['pass_outcome'].isin(['Incomplete', 'Pass Offside'])]

pitch.arrows(pases_completos['x'], pases_completos['y'], pases_completos['endx'], pases_completos['endy'], ax=ax, label='Pases Completos', color='green', width=2.5 )
pitch.arrows(pases_incompletos['x'], pases_incompletos['y'], pases_incompletos['endx'], pases_incompletos['endy'], ax=ax, label='Pases Incompletos', color='red', width=2.5 )

plt.legend()
```

```python
bundesliga_ids = list(bundesliga[(bundesliga['home_team'] == 'Bayer Leverkusen') | (bundesliga['away_team'] == 'Bayer Leverkusen')].match_id.values)

```

```python
tiros = pd.DataFrame()
for i in bundesliga_ids:
    df = sb.events(match_id=i)
    df = df[df['type'] == 'Shot']
    tiros = pd.concat([tiros, df], ignore_index=True)
```

```python
tiros[['x', 'y']] = tiros.location.apply(pd.Series)
tiros[['x', 'y', 'location']]
```

```python
tiros.possession_team.value_counts()
```

```python
tiros['shot_team'] = np.where(tiros['possession_team'] == 'Bayer Leverkusen', 'Bayer Leverkusen', 'No Bayer Leverkusen' )
tiros.shot_team.value_counts()
```

```python
tiros_Leverkusen = tiros[tiros['shot_team'] == 'Bayer Leverkusen' ]
tiros_Leverkusen.shape #622 tiros como arriba
```

```python
fig, ax = plt.subplots(figsize=(16,9))
pitch = VerticalPitch(
    pitch_type='statsbomb',
    axis=True,
    label=True,
    tick=True,
    goal_type='box',
    half=True,
    
)

pitch.draw(ax=ax)

pitch.scatter(tiros_Leverkusen.x, tiros_Leverkusen.y, ax=ax, s=tiros_Leverkusen.shot_statsbomb_xg*400 , alpha=.6, ec='black', label='Tiros' )
# s= (parametro que te permite agregar una nueva variable al grafico, en este caso los xg)
# alpha (transparencia, lo que ayuda a visualizar los pintos que se superponen)
# ec  (es para darle contorno al punto) 
# label (ademas de ser una etiqueta para mostrat en el propio grafico, funciona como en css, se puede filtrar por los labels)
```

```python
tiros_Leverkusen.shot_outcome.value_counts()
```

```python
goles_leverkusen = tiros_Leverkusen[tiros_Leverkusen.shot_outcome == 'Goal']
no_goles_leverkusen = tiros_Leverkusen[tiros_Leverkusen.shot_outcome != 'Goal']
no_goles_leverkusen.shape , goles_leverkusen.shape
#88 filas, lo que corresponde a los 88 goles en la temporada
```

```python
fig, ax = plt.subplots(figsize=(16,9))
pitch = VerticalPitch(
    pitch_type='statsbomb',
    axis=True,
    label=True,
    tick=True,
    goal_type='box',
    half=True,
    
)

pitch.draw(ax=ax)


pitch.scatter(goles_leverkusen.x, goles_leverkusen.y, ax=ax, s=goles_leverkusen.shot_statsbomb_xg*500 , alpha=.6, ec='black', label='Tiros terminados en gol', zorder=.99 )
pitch.scatter(no_goles_leverkusen.x, no_goles_leverkusen.y, ax=ax, s=no_goles_leverkusen.shot_statsbomb_xg*400 , alpha=.6, ec='black', label='Tiros errados', zorder=.0 )
#zorden (Orden en el eje Z, basicamente va de 0 a 1, dandole prioridad de visualizacion en la superposicion, siendo .99 los que estan mas arriba)

plt.legend()
```
